#### Refactoring

```python
def analysis(df, group, agg, file='../data/result.csv'):
    analyzed = df.groupby(by=group)[agg].mean()
    analyzed.to_csv(file)
    return analyzed
```

```python
df_vehicles_analyzed = analysis(df_vehicles, 'Model', 'Fuel Barrels/Year', '../data/resultFuel.csv')
df_vehicles_analyzed
```
